```python
import os
import pandas as pd
import numpy as np
import plotly.express as px


current_directory = os.getcwd()
```

```python
daily_data_df = pd.read_csv(f"{current_directory}\\worldometer_coronavirus_daily_data.csv")
daily_data_df_copy = daily_data_df.copy()

daily_data_df_copy
```

```python
summary_data_df = pd.read_csv(f"{current_directory}\\worldometer_coronavirus_summary_data.csv")

summary_data_df_copy = summary_data_df.copy()
summary_data_df_copy
```

```python
summary_data_df_copy.dtypes
```

```python
summary_data_df_copy.drop(columns=['serious_or_critical', 'active_cases', 'total_deaths_per_1m_population', 'total_cases_per_1m_population', 'total_tests_per_1m_population'], inplace=True)

summary_data_df_copy['continent'].unique()
```

```python
summary_data_df_copy.fillna(0, inplace=True)

summary_data_df_copy[summary_data_df_copy['continent'] == 'Asia']
```

```python
# array(['Asia', 'Europe', 'Africa', 'North America', 'South America',
#        'Australia/Oceania'], dtype=object)
summary_criteria = ['population', 'total_confirmed']
sorted_summary_asia = summary_data_df_copy[summary_data_df_copy['continent'] == 'Asia'].head(5).sort_values(by=summary_criteria, ascending=False)
sorted_summary_europe = summary_data_df_copy[summary_data_df_copy['continent'] == 'Europe'].head(5).sort_values(by='total_confirmed', ascending=False)
sorted_summary_africa = summary_data_df_copy[summary_data_df_copy['continent'] == 'Africa'].head(5).sort_values(by='total_confirmed', ascending=False)
sorted_summary_na = summary_data_df_copy[summary_data_df_copy['continent'] == 'North America'].head(5).sort_values(by='total_confirmed', ascending=False)
sorted_summary_sa = summary_data_df_copy[summary_data_df_copy['continent'] == 'South America'].head(5).sort_values(by='total_confirmed', ascending=False)
sorted_summary_au = summary_data_df_copy[summary_data_df_copy['continent'] == 'Australia/Oceania'].head(5).sort_values(by='total_confirmed', ascending=False)
```

```python
sorted_summary_asia

fig = px.bar(
    sorted_summary_asia,
    x ='country',
    y = 'total_confirmed',
    color='continent'
)

fig.show()
```
